```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
print('imported')
```

```python
# load the yolo model
model=YOLO('yolov8n.pt') #n-nano
```

```python
print(model)
```

```python
image_path='/content/animals.jpg'
```

```python
# run the object detection
result=model(image_path)
```

```python
# display results
for r in result:
  im_array=r.plot()
  # convert BGR to RGB
  im_rgb=cv2.cvtColor(im_array,cv2.COLOR_BGR2RGB)
  plt.figure(figsize=(10,6))
  plt.imshow(im_rgb)
  plt.axis('off')
  plt.show()
```
